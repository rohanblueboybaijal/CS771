```python
# Perceptron Model
class Perceptron:
    def __init__(self):
        # Model is described by the weight
        self.w = None
    
    # Prediction function
    def predict(self, X):
        # y_hat stores the prediction result of prediction for each example
        y_hat = []
        N = X.shape[0] # no. of rows/examples
        for x in X:
            # We follow the rule given in lecture
            if np.dot(self.w, x) >=0 :
                y_hat.append(1)
            else:
                y_hat.append(-1)
    
    # Function to compute loss
    def compute_loss(self, X, Y):
        loss = 0
        for n, xn in enumerate(X):
            if np.dot(self.w, xn)*Y[n] < 0:
                loss += 1
        # Loss is taken to be the ratio of mistakes made
        return loss / X.shape[0]
    
    # NOTE : We are not putting any break condition here since we don;t know if the data is linearly separable
    def fit(self, X, Y, epochs=5000, eta=1):
        # seed to ensure reproducibility
        np.random.seed(0)
        
        # history keeps track of loss after each iteration to visualize loss vs iteration and check for covnergence
        loss_history = []
        # random initialization
        self.w = np.random.randint(0, 10, X.shape[1])
        
        for t in range(epochs):
            # Randomly sampling an example
            index = np.random.choice(X.shape[0])
            y_n = Y[index]
            x_n = X[index]
            
            # Update weights if mistake is made
            if y_n * np.dot(self.w, x_n) < 0:
                self.w = self.w + eta*y_n*x_n
            
            # Compute and store loss
            loss =self.compute_loss(X, Y)
            loss_history.append(loss)
            
        return loss_history
    
    # Function to compute F1 score
    def compute_F1(self,X, Y):
        TN = 0 # True negative
        TP = 0 # True Positive
        FN = 0 # False Negative
        FP = 0 # False Positive
        
        # Incerement these values based on the rules
        for n, xn in enumerate(X):
            # +1 is considered fake note
            if np.dot(self.w, xn) >= 0:
                if Y[n] > 0:
                    TP += 1
                else:
                    FP += 1
            else:
                if Y[n] > 0:
                    FN += 1
                else:
                    TN += 1
                    
        # Precision calculation
        precision = TP/(TP + FP)
        # Recall Calculation
        recall = TP/(TP + FN)
        
        # F1 = 2PR/(P+R)
        F1 = 2*precision*recall/(precision + recall)
        return precision, recall, F1
        
```

```python
# Function to split data into K folds for K-fold Cross Validation
def CrossValidationSplit(df, K):
    # Seed for reproducibility
    np.random.seed(0)
    
    # number of samples in each fold
    fold_size = int(df.shape[0]/K)
    
    # Storing each split ; making copies so that data isn't affected by memory management
    splits = list()
    df = df.sample(frac=1)
    df_copy = df.copy(deep=True)
    
    # Create folds
    for i in range(K):
        fold = list()
        # Keep adding example until fold is of required size; drop elements so that an element isn't repeated
        while(len(fold) < fold_size):
            index = np.random.randint(df_copy.shape[0])
            index = df_copy.index[index]
            fold.append(df_copy.loc[index].values.tolist())
            df_copy.drop(index, inplace=True)
        
        splits.append(np.array(fold))
    
    return np.array(splits)
```

```python
# Function to perform K-fold cross validation
def CrossValidation(df, K, lr):
    df_copy = df.copy(deep=True)
    
    # Get K splits 
    split = CrossValidationSplit(df, K)
    
    # Loss keeps track of the final loss on for each fold as test 
    losses = list()
    # lh keeps track of the training loss history for each fold
    lh = []
    
    for k in range(K):
#         print("Fold " + str(k+1))
        split_copy = np.copy(split)
        # Use one fld as test data
        test_set = split_copy[k]
        X_test = test_set[:,:-1] # remove last column since that is the result
        Y_test = test_set[:,-1] # last column
        
        # All the other folds become training data
        train_set = np.delete(split_copy, k, axis=0).reshape(-1, 5)
        X_train = train_set[:,:-1] # remove last column
        Y_train = train_set[:,-1] # last column is the result
        
        # Adding 1 in order to incorporate bias term of weights
        train_one_col = np.ones((X_train.shape[0],1))
        X_train = np.hstack((X_train, train_one_col))
        
        test_one_col = np.ones((X_test.shape[0],1))
        X_test = np.hstack((X_test, test_one_col))
        
        # Need to have +1/-1 label
        Y_test[Y_test==0] = -1
        Y_train[Y_train==0] = -1
        
        # Initialize Perceptron model
        perceptron = Perceptron()
        # Train Model
        hist = perceptron.fit(np.array(X_train), np.array(Y_train), epochs=5000, eta=lr)
        
        # Compute final loss on the fold
        loss = perceptron.compute_loss(np.array(X_test), np.array(Y_test))
        losses.append(loss)
        
        # Stores history of training loss
        lh.append(hist)
    
    losses = np.array(losses)
    # Report average loss
    return np.mean(losses), lh
        
```

# 2) Markov Chain Monte Carlo

## a) Computing Posterior Distribution
We are given N data samples $t_i$~$\mathcal{N}(\mu, \sigma)$.    

We are also given a prior on our $\mu$ ~ $\mathcal{N}(\mu_0, \sigma_0)$.  
Now the likelihood is 
$P(\mu \mid t_1, t_2 ... t_N) \propto P(t_1, t_2, ... t_N \mid \mu)P(\mu)$  
Since the observations are independent, $P(t_1, t_2, \dots, t_N \mid \mu) = P(t_1 \mid \mu)P(t_2 \mid \mu) \dots P(t_N \mid \mu)$  

Therefore   
$P(\mu \mid t_1, t_2 \dots t_N) \propto \frac{1}{\sqrt{2\pi\sigma_0^2}} exp(\frac{-(\mu - \mu_0)^2}{2\sigma_0^2}) \times \prod_{i=1}^N \frac{1}{\sqrt{2\pi\sigma_0^2}} exp(\frac{-(t_i - \mu)^2}{2\sigma^2})$   
$= \frac{1}{(2\pi)^{\frac{N+1}{2}}\sigma_0\sigma^N}exp\biggl( \frac{-\mu^2 + 2\mu\mu_0 -\mu_0^2}{2\sigma_0^2} - \sum_{i=1}^N \frac{t_i^2 -2\mu t_i + \mu^2}{2\sigma^2} \biggr)$  
We now focus on the exponential term only and use the square completion technique to determine new Gaussian parameters(Posterior) 

The term inside can be worked out as :  
$\frac{-1}{2\sigma^2}(\sum_{i=1}^N t_i^2 + N\mu^2 - 2 \sum_{i=1}^N t_i \mu) - \frac{1}{2\sigma_0^2}(\mu^2 + \mu_0^2 - 2\mu \mu_0)$  

$= \frac{-1}{2} \biggl( \mu^2( \frac{N}{\sigma^2} + \frac{1}{\sigma_0^2}) + \mu(\frac{-2 \sum_{i=1}^N t_i}{\sigma^2} - \frac{2\mu_0}{\mu_0^2})  + \frac{\sum_{i=1}^N t_i^2}{\sigma^2} + \frac{\mu_0^2}{\sigma_0^2} \biggr)$  

Suppose the new mean and variance is $\mu_N$ and $\sigma_N$.  
We know that in the exp, the term is $\frac{-1}{2\sigma_N^2}(\mu - \mu_N)^2$.  We can now match to get $\mu_N$ and $\sigma_N$

By making comparisons, we see that  $\frac{1}{\sigma_N^2} = \frac{N}{\sigma^2} + \frac{1}{\sigma_0^2}$  
Therefore 
$\sigma_N^2 = \frac{\sigma^2 \sigma_0^2}{N \sigma_0^2 + \sigma^2} $  

For $\mu_N$, we compare the terms $\frac{-\mu \mu_N}{\sigma_N^2} $ with $ -\mu \biggl( \frac{\sum_{i=1}^N}{\sigma^2} + \frac{\mu_0}{\sigma_0^2} \biggr)$

Also, $\mu_N = \sigma_N^2 \biggl( \frac{\sum_{i=1}^N}{\sigma^2} + \frac{\mu_0}{\sigma_0^2} \biggr)$

## Computing Likelihood from data
We have $N$ samples $t_i$ from a normal distribution.  They are assumed to be iid and so the likelihood $L(x_1, x_2, \dots, x_N)$ can be seen as a product of independent probabilities.  

$L(t_1, t_2, \dots, t_N \mid \mu) = P(t_1, t_2, \dots, t_N \mid \mu)$  

$ =\prod_{i=1}^N \frac{1}{\sqrt{2\pi\sigma_0^2}} exp(\frac{-(t_i - \mu)^2}{2\sigma^2})$  
From the multiplication and comparison of terms (using completing the square technique) we find that :  
$\hat{\mu} = \frac{\sum_{i=1}^Nt_i}{N}$  
$\hat{\sigma} = \frac{\sigma}{\sqrt{N}}$

```python
# Custom Gaussian function for plots and to compute normal pdf
# mu is the mean
# sigma is the standard deviation of Gaussian
def gaussian(x, mu, sigma):
    return np.exp(-np.square(x-mu)/(2*sigma*sigma))/(np.sqrt(2*np.pi)*sigma)
```

```python
np.random.seed(0)

N = 50 # number of samples
variance = 25 # variance of the samples
variance_0 = 25 # variance of the prior

variance_1 = variance*variance_0/(N*variance_0 + variance) # updated variance of mu (posterior)

# Generating 50 random samples from Gaussian distribution with mean=10 and variance=25
t = np.random.normal(10, 5, size=(1,N))
t_sum = np.sum(t)
# Print sum and empirircal mean
print(t_sum, t.mean())

# Prior mean = 10
mu_0 = 25
mu_1 = variance_1*((mu_0/variance_0) + t_sum/variance) # Posterior mean of mu

print(mu_1, variance_1)
```

```python
# Plotting the Guassian pdfs on (0,40)
# NOTE : The likelihood has been scaled for convenience

# Parameters of the Likelihood
mean = t_sum/N; std = 5; variance = np.square(std)/N
x = np.arange(0,40,.01)
f = gaussian(x, mean, np.sqrt(variance))
plt.plot(x,f, label="Likelihood")

# Parameters of the prior 
mean = 25; std = 5; variance = np.square(std)
x = np.arange(0,40,.01)
f = gaussian(x, mean, std)
plt.plot(x,f, label="Prior", color="g")

# Parameters of the posterior
mean = mu_1; std = np.sqrt(variance_1)
x = np.arange(0,40,.01)
f = gaussian(x, mean, std)
plt.plot(x,f, label="Posterior", color="r")

plt.ylabel('gaussian distribution')
plt.legend()
plt.show()
```

Defining the transition model for $\mu_{current}$ to $\mu_{new}$  
$Q(\mu_{new} \mid \mu_{current}) = \mathcal{N}(\mu = \mu_{current} , \sigma')$

```python
# Used for generating new proposal from current value : Random walk
def transition_model(mu_current, sigma):
    # Transition model is again Gaussian
    return np.random.normal(mu_current,sigma,1)
```

```python
# This computes the prior for given mu. In our case prior mean = 25 and prior variance = 25
def compute_prior(mu, mu_prior=25, sigma_prior=5):
    return gaussian(mu, mu_prior, sigma_prior)
```

```python
# We will use the log likelihood instead of the likelihood in order to ensure numerical stability
def log_likelihood(mu,sigma,data):
    return np.sum(np.log(gaussian(data, mu, sigma)))
```

```python
def acceptance(x, x_new):
    if x_new > x:
        return True
    else:
        accept = np.random.uniform(0,1)
        # Since we use log likelihood, we need to use the exponentiation function
        return accept < np.exp(x_new - x)
```

## Deal with Acceptance Probability  

Notice that the acceptance probability $\alpha$ deals with ratios of likelihoods  

$\alpha = \frac{Prior(\mu_{new})L(D \mid \mu_{new})}{Prior(\mu_{old})L(D \mid \mu_{old})}$ if this term is less than 1, otherwise $\alpha=1$   
where D is the data  
Now, notice that the $L$ involves a lot of multiplications of Gaussians in this case. This can lead to some very very small values which might in turn lead to errors.  

INstead of usig the ratios as is, we use the log-likelihood.  

We are interested in the term : $\log(L(D \mid \mu_{new})) + \log(Prior(\mu_{new}))  - \log(L(D \mid \mu_{old})) - \log(Prior(\mu_{old})) $  
If this term is greater than 0, then we simply put $\alpha = 1$.  
Otherwise, we raise $e$ to this term and return that as $\alpha$  
Hence, both are equivalent and using log likelihood can lead to numerical stability as we scale up the algorithm

```python
landmarks = [1000, 5000, 10000, 20000, 50000] # We are going to check for speed of convergence by plotting the histogram in intermediate iterations

# Posterior pdf
mean = mu_1; std = np.sqrt(variance_1)
x = np.arange(0,40,.01)
posterior_dist = gaussian(x, mean, std)


# MH algorithm
def MetropolisHastings(data=t, mu_prior=25, sigma_prior=5, sigma=5, param_init=0, sigma_transition=1, iterations=1000):
    # data : Given samples
    # mu_prior : mean of the prior distribution 
    # sigma_prior : standard deviation of the prior distribution
    # sigma : variance of the data being sampled : it is assumed to be constant
    # param_init : initial value of mu we start the algorithm from
    # sigma_transition : defines the variance of the transition model which is again Gaussian
    
    np.random.seed(0) # Set seed for reproducibility
    
#     fig,axs = plt.subplots(1,4, figsize=(25,5))
    
    mu = param_init # Set mu to initial value
    accepted = [] # We will plot histogram using accepted proposals only
    rejected = []
    
    # Start sampling
    for i in range(iterations):
        # Generate new value of mu (Random walk)
        mu_new = transition_model(mu, sigma_transition)
        
        # Calculate log likelihoods
        mu_lik = log_likelihood(mu, sigma, data)
        mu_new_lik = log_likelihood(mu_new, sigma, data)
        
        # Compute the log terms for the acceptance probability
        x = mu_lik + np.log(compute_prior(mu, mu_prior, sigma_prior))
        x_new = mu_new_lik + np.log(compute_prior(mu_new, mu_prior, sigma_prior))
        
        # Randomly accept proposal based on ration of likelihoods
        if(acceptance(x, x_new)):
            mu = mu_new
            accepted.append(mu)
        else:
            rejected.append(mu_new)
    
    return np.array(accepted), np.array(rejected)
```

```python
acc, rej = MetropolisHastings(data=t, mu_prior=25, sigma_prior=5, sigma=5, param_init=0, sigma_transition=1, iterations=50000)
print("Mean of accepted proposals : ", acc.mean())

# Plot the histogram
fig, axs = plt.subplots(1, 1, figsize =(10, 7), tight_layout = True)
axs.hist(acc.reshape(-1,), np.arange(0,25,0.1), density=True)

# Parameters of the posterior
mean = mu_1; std = np.sqrt(variance_1)
x = np.arange(0,20,.01)
f = gaussian(x, mean, std)
# Plot the posterior
axs.plot(x,f, label="Posterior", color="r")

plt.legend()
fig.show()
```

## Results
We notice that all transition models fail to give good converage in 1000 iterations.  
However, $\sigma < 1.5$ starts giving decent coverage by 10000 iterations but seem to converge properly only till 20000 iterations. 

A special case is $\sigma = 0.75$ which is very close to the posterior value. That has good quality and converges in around 5000 iterations only.  

For higher values of $\sigma$ we get bad coverage. The area under the analytic posterior is not covered well and there are significant samples. The part around the computed mean is empty since the transition model takes very big jumps and can't converge around the mean properly.  

On the other hand, for smaller values of $\sigma$, we notice a lot of overshoot around the posterior mean. This is because the algorithm doesn't get a chance to explore too much and once it reaches the mean, it stays around it a lot because of the small steps.  

However, we notice that $\sigma=0.75$ takes care of the overshoot much earlier than in the case of smaller values of $\sigma$
